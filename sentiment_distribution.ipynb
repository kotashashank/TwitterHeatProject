```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
tweets = pd.read_csv('manual_filtered_travis_tweets_with_blockgroups_with_scores_and_detailed_extra_sentiment.csv')

```

```python
plt.hist(tweets['absolute_score'], density = True, facecolor = '#ffb100', edgecolor='#000', linewidth=0.5)
plt.title('Absolute')
tweets['absolute_score'].plot(kind = "kde")
plt.grid(False)
plt.style.use('seaborn-whitegrid') # nice and clean grid
plt.xlim(-1, 1)
```

```python
plt.hist(tweets['relative_proportional_score'], density = True, facecolor = '#ffb100', edgecolor='#000', linewidth=0.5)
plt.title('Relative Proportional')
tweets['relative_proportional_score'].plot(kind = "kde")
plt.grid(False)
plt.style.use('seaborn-whitegrid') # nice and clean grid
plt.xlim(-1, 1)
```

```python
no_na_logit = tweets[~tweets['logit_score'].isna()]
print(tweets['logit_score'])
no_na_logit = no_na_logit['logit_score']
plt.hist(no_na_logit, density = True, facecolor = '#ffb100', edgecolor='#000', linewidth=0.5)
plt.title('Logit Scale')

no_na_logit.plot(kind = "kde")
plt.grid(False)
plt.style.use('seaborn-whitegrid') # nice and clean grid
plt.xlim(-1, 1)
```

```python
from scipy.stats import skew, kurtosis

# skewness = 
no_na_rel = tweets[~tweets['relative_proportional_score'].isna()]
no_na_rel = no_na_rel['relative_proportional_score']
print("Absolute skewness score: ", skew(tweets['absolute_score']))
print("Relative Proportion skewness score: ", skew(no_na_rel))
print("Logit Scale skewness score: ", skew(tweets['logit_score']))
```

```python
print("Absolute kurtosis score: ", kurtosis(tweets['absolute_score']))
print("Relative Proportion kurtosis score: ", kurtosis(no_na_rel))
print("Logit Scale kurtosis score: ", kurtosis(tweets['logit_score']))
```
